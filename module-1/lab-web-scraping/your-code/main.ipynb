####  20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe

````python
# This is the url you will scrape in this exercise
URL_EARTHQUAKE = 'https://www.emsc-csem.org/Earthquake/'
````

````python
# data is where the final information will be store
#Every td label has a datum, some positions are useful (3,4,6 and 11)
#Thus, we have to jump 13 positions till next useful datums
#jump allow to jump till the correct index, and them record the datum

data = []
temp = []
jump = 0
    
response = requests.get(URL_EARTHQUAKE)
soup = BeautifulSoup(response.text, 'html.parser')
container = soup.find('tbody',{'id':'tbody'})

readings = container.find_all('td') 

for index,reading in enumerate(readings):
    if(index == 3+jump):
        temp.append(reading.text[10:20])
        temp.append(reading.text[23:31])
    if(index == 4+jump):
        temp.append(reading.text.replace('\xa0',''))
    if(index == 6+jump):
        temp.append(reading.text.replace('\xa0',''))
    if(index == 11+jump):   
        temp.append(reading.text.replace('\xa0',''))
        jump+=13
        data.append(temp[0:5])
        temp = []

earthquake_df = pd.DataFrame(data, columns=['Date', 'Time', 'Lat', 'Long', 'Region'])
earthquake_df.head(20)
````

#### Display the date, days, title, city, country of next 25 hackathon events as a Pandas dataframe table

````python
# This is the url you will scrape in this exercise
URL_HACKATHON ='https://hackevents.co/hackathons'
URL_CONSTRUCTOR = 'https://hackevents.co'
````

````python
# All nesserary information about events should be gathered from the links of the events
# Thus, we find the lines that start with "/event", them we get the link stored in the label
# href in order to build the correct url adding the URL_CONSTRUCTOR (the root url)
# Them we have to interate through the correct links, and get the data

data = []
links = []
    
response = requests.get(URL_HACKATHON)
soup = BeautifulSoup(response.text, 'html.parser')
containers = soup.find_all('a')

# loop to get the links for all events
for container in containers:
    if (container.get('href')[0:6] == '/event'):
        links.append(URL_CONSTRUCTOR + container.get('href'))

# loop to scrap every event page
for link in links:
    response = requests.get(link)
    soup = BeautifulSoup(response.text, 'html.parser')
    container = soup.find('div',{'class','row'})

    #title
    title = container.find('h2').text
    
    #the rest of the datums
    datums = container.find_all('p')
    
    city = datums[0].text.replace('\u2003 ','').split(',')[0]        
    country = datums[0].text.replace('\u2003 ','').split(',')[2] 
    start_date = datums[1].text.split(' ')[-1]
    start_day = int(start_date.split('/')[1])

    end_date = datums[2].text.split(' ')[-1] 
    end_day = int(end_date.split('/')[1])
    gape = end_day - start_day
    days = [start_day+day for day in range(gape+1)]
    
    data.append([start_day,days,title,city,country])

hackathons_df = pd.DataFrame(data, columns=['start_day', 'days', 'title', 'city', 'country'])
hackathons_df.head(25)
````

#### IMDB's Top 250 data (movie name, Initial release, director name and stars) as a pandas dataframe

````python
# This is the url you will scrape in this exercise 
url = 'https://www.imdb.com/chart/top'
````

````python
# your code    
def get_movie_data(soup):
    movie_data = list()
    movies_list = list()
    count = 0
    
    print("Scrapping ",url,"page")
    
    movies = soup.find_all("td", {"class": "titleColumn"})
    
    for movie in movies:
        movie_data = None
        try:
            movie_data = get_movie_text(movie)
            
        except Exception as e:
            continue
            #ignore if there is any loading or movie error

        if movie_data:
            movies_list.extend(movie_data)
            #forcing the printing of the "." by flushing the buffer. One point for every input
            print(".", end="")
            sys.stdout.flush()
            count +=1
        
        if count == 250: #Maximum of 250 movies scrapped
            break
    
    return movies_list

def get_movie_text(movie):
    movie_text = list()
    
    title = str((movie.find("a")).text).strip()
    release = str((movie.find('span',{'class','secondaryInfo'})).text).strip()
    director = str(movie.find("a")['title']).strip().split("(dir.), ")[0]
    stars = str(movie.find("a")['title']).strip().split("(dir.), ")[1]

    movie_text.append([title,release,director,stars])
    
    return movie_text  

def movies_scrap():    
    try:
        response = requests.get(url) 
    except Exception as e:
        print(repr(e))
        sys.exit(1)
    
    if response.status_code != 200:
        print("Non success status code returned "+str(response.status_code))
        sys.exit(1)

    soup = BeautifulSoup(response.text, 'lxml')

    if soup.find("a", {"class": "titleColumn"}):
        print("\n\n Error: Invalid page.")
        sys.exit(1)

    movies_list = get_movie_data(soup)

    df = pd.DataFrame(movies_list,columns=['TITLE','RELEASE','DIRECTOR','STARS'])
    df.index += 1
    
    return df
````

````python
#your code
# your code    
def get_movie_data(soup):
    movie_data = list()
    movies_list = list()
    
    print("Scrapping ",url,"page")
    
    #get only the meta data of 10 movies
    movies = soup.find_all("td", {"class": "titleColumn"})[:10]
    
    for movie in movies:
        movie_data = None
        try:
            movie_data = get_movie_text(movie)
                
        except Exception as e:
            continue
            #ignore if there is any loading or movie error

        if movie_data:
            movies_list.extend(movie_data)
            #forcing the printing of the "." by flushing the buffer. One point for every input
            print(".", end="")
            sys.stdout.flush()
        
    return movies_list



def get_movie_text(movie):
    movie_text = list()
    
    title = str((movie.find("a")).text)
    release = str((movie.find('span',{'class','secondaryInfo'})).text).strip()

    href= "http://www.imdb.com" + movie.find("a")['href']
    brief= brief_scrap(href)
    movie_text.append([title,release,brief])

    return movie_text  

def movies_scrap():    
    try:
        response = requests.get(url) 
    except Exception as e:
        print(repr(e))
        sys.exit(1)
    
    if response.status_code != 200:
        print("Non success status code returned "+str(response.status_code))
        sys.exit(1)

    soup = BeautifulSoup(response.text, 'lxml')

    if soup.find("a", {"class": "titleColumn"}):
        print("\n\n Error: Invalid page.")
        sys.exit(1)

    movies_list = get_movie_data(soup)

    df = pd.DataFrame(movies_list,columns=['TITLE','RELEASE','BRIEF'])
    df.index += 1
    
    return df


def brief_scrap(url_brief):    
    try:
        response = requests.get(url_brief) 
        
    except Exception as e:
        print(repr(e))
        sys.exit(1)
    
    if response.status_code != 200:
        print("Non success status code returned "+str(response.status_code))
        sys.exit(1)

    soup_brief = BeautifulSoup(response.text, 'lxml')

    if soup_brief.find("a", {"class": "titleColumn"}):
        print("\n\n Error: Invalid page.")
        sys.exit(1)

    movie_brief = get_movie_brief(soup_brief)
    
    return movie_brief


def get_movie_brief(soup_brief):
    try:
        brief = soup_brief.find("div",{'class',"summary_text"}).text
    except Exception as e:
        #continue  
        pass
    
    return brief
````
